```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gurugram_weather_2023.csv")

df = df[['datetime', 'temp', 'tempmax', 'tempmin', 'precip', 'humidity']]


df = df.dropna()                                   
df['datetime'] = pd.to_datetime(df['datetime'])    
df = df.set_index('datetime')

print(f"FINAL SUCCESS! {len(df)} days loaded perfectly")

plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
plt.plot(df.index, df['temp'], color='red', linewidth=1.2)
plt.title('Daily Average Temperature  Gurugram 2023', fontsize=14)
plt.ylabel('Temperature (C)')
plt.grid(alpha=0.3)

plt.subplot(2,2,2)
monthly_rain = df['precip'].resample('M').sum()
plt.bar(monthly_rain.index.strftime('%b'), monthly_rain, color='skyblue', edgecolor='navy')
plt.title('Monthly Total Rainfall', fontsize=14)
plt.ylabel('Rainfall (mm)')

plt.subplot(2,2,3)
plt.scatter(df['temp'], df['humidity'], alpha=0.6, color='purple')
plt.title('Temperature vs Humidity', fontsize=14)
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')
plt.grid(alpha=0.3)

plt.subplot(2,2,4)
plt.plot(df.index, df['humidity'], color='green', linewidth=1.2)
plt.title('Daily Humidity Trend', fontsize=14)
plt.ylabel('Humidity (%)')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig("gurugram_final_plots.png", dpi=300, bbox_inches='tight')
plt.show()

df.to_csv("cleaned_gurugram_weather.csv")
print("COMPLETED 100%! Check C:\\Users\\KENISHA for:")
print("    gurugram_final_plots.png")
print("    cleaned_gurugram_weather.csv")
```
